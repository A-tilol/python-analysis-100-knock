## 32 clusterize customers

```python
customer_clustering = customer[["mean", "median", "max", "min", "membership_period"]]
customer_clustering.head()
```

### normalize

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()

customer_clustering_sc = sc.fit_transform(customer_clustering)
print(customer_clustering_sc[:5])
```

### K-means

```python
kmeans = KMeans(n_clusters=4, random_state=0)
clusters = kmeans.fit(customer_clustering_sc)
customer_clustering["cluster"] = clusters.labels_
print(customer_clustering["cluster"].unique())
customer_clustering.head()
```

## 36 prepare for usage count prediction

```python
uselog.head()
```

```python
uselog.usedate = pd.to_datetime(uselog.usedate)
uselog["yearmonth"] = uselog.usedate.dt.strftime("%Y%m")
uselog_months = uselog.groupby(["customer_id", "yearmonth"], as_index=False).count()
uselog_months.rename(columns={"log_id": "count"}, inplace=True)
del uselog_months["usedate"]
uselog_months.head()
```

```python
yearmonths = list(uselog_months.yearmonth.unique())

preict_data = pd.DataFrame()
for i in range(6, len(yearmonths)):
    tmp = uselog_months[uselog_months.yearmonth == yearmonths[i]]
    tmp.rename(columns={"count": "count_pred"}, inplace=True)
    
    for j in range(1, 7):
        tmp_before = uselog_months[uselog_months.yearmonth == yearmonths[i-j]]
        del tmp_before["yearmonth"]
        tmp_before.rename(columns={"count": f"count_{j-1}"}, inplace=True)
        
        tmp = pd.merge(tmp, tmp_before, on="customer_id", how="left")
        
    preict_data = pd.concat([preict_data, tmp], ignore_index=True)
    
preict_data.head()
```

```python
print(preict_data.shape)
preict_data2 = preict_data.dropna()
print(preict_data2.shape)
preict_data2.head()
```

## 36 add features

```python
preict_data2 = pd.merge(preict_data2, customer[["customer_id", "start_date"]], on="customer_id", how="left")
preict_data2.start_date = pd.to_datetime(preict_data2.start_date)
preict_data2.head()
```

```python
preict_data2["now_date"] = pd.to_datetime(preict_data2.yearmonth, format="%Y%m")
preict_data2.head()
```

```python
from dateutil.relativedelta import relativedelta

periods = []
for i in range(len(preict_data2)):
    delta = relativedelta(preict_data2.now_date[i], preict_data2.start_date[i])
    periods.append(delta.years * 12 + delta.months)
preict_data2["period"] = periods

preict_data2.head()
```

## 38 predict usage count

```python
print(preict_data2.shape)
preict_data3 = preict_data2[preict_data2.start_date >= pd.to_datetime("20180401")]
print(preict_data3.shape)
preict_data3.head()
```

```python
from sklearn import linear_model
import sklearn.model_selection

model = linear_model.LinearRegression()
X = preict_data3[["count_0", "count_1", "count_2", "count_3", "count_4", "count_5", "period"]]
y = preict_data3.count_pred

X_train, X_test, y_train, y_test = sklearn.model_selection.train_test_split(X, y)

model.fit(X_train, y_train)
```

```python
print(model.score(X_train, y_train))
print(model.score(X_test, y_test))
```
